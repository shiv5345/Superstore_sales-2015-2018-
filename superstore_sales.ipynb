##### create new column

```python
df.drop_duplicates(inplace = True)
```

```python
df.head()
```

```python
df['Sales'] = df['Sales'].round(2)
df.head()
```

## Removing Outliers by the basis of Total Sales

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_bound = Q1-1.5*IQR
lower_bound
```

```python
upper_bound = Q3 + 1.5*IQR
upper_bound
```

```python
df_no_outliers = df[(df['Sales']>=lower_bound) & (df['Sales']<=upper_bound)]
```

```python
print("Before:", df.shape)
print("After:", df_no_outliers.shape)
```

```python
df=df_no_outliers
```
